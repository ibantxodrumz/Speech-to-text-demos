```python
""""I need to to build a dummy dataset. it needs to have an a time colum where you start at oo:oo and finish at 23:40.
Then an ID column with 1 single ID called VN1. After this we need a column called Expected with numbers from 50 to 100
with two decimals. another column called Obtained is the same as expected, same type of number. Finally Another one called Difference 
it shoudl be (Expected - Obtained)""" 
```

```python
# Generate the time column with 30-minute intervals from 00:00 to 23:40
time = [f"{str(h).zfill(2)}:{str(m).zfill(2)}" for h in range(24) for m in range(0, 60, 30)]

# Use a constant ID 'VN1' for all rows
id_column = ['VN1'] * len(time)

# Generate the 'Expected' and 'Obtained' columns with random values from 50 to 100 (two decimals)
expected = [round(random.uniform(50, 100), 2) for _ in range(len(time))]
obtained = [round(random.uniform(50, 100), 2) for _ in range(len(time))]

# Calculate the 'Difference' column as the absolute difference between 'Expected' and 'Obtained'
difference = [abs(exp - obs) for exp, obs in zip(expected, obtained)]

# Create a DataFrame with the generated data
df = pd.DataFrame({'Time': time, 'ID': id_column, 'Expected': expected, 'Obtained': obtained, 'Difference': difference})
```

```python
# Import necessary libraries
import matplotlib.pyplot as plt

# Filter the DataFrame for the 'VN1' ID
vn1_data = df[df['ID'] == 'VN1']

# Plot 'Expected' and 'Obtained' columns against 'Time'
plt.figure(figsize=(12, 6))
plt.plot(vn1_data['Time'], vn1_data['Expected'], label='Expected', marker='o')
plt.plot(vn1_data['Time'], vn1_data['Obtained'], label='Obtained', marker='x')

# Customize the plot
plt.title('Expected vs Obtained for VN1')
plt.xlabel('Time')
plt.ylabel('Values')
plt.xticks(rotation=45)  # Rotate x-axis labels for readability
plt.legend()
plt.grid(True)

# Display the plot
plt.show()

# Create a new column 'difference_sign' to store the difference between 'Expected' and 'Obtained' while preserving the sign
df['difference_sign'] = df['Expected'] - df['Obtained']
df.head(10)

# Calculate the 'Error' as the ratio of 'Difference' to 'Expected'
df['Error'] = df['difference_sign'] / df['Expected']

# Calculate the 'Error_sign' as the ratio of 'difference_sign' to 'Expected'
df['Error_sign'] = df['difference_sign'] / df['Expected']
df.head(10)

# Calculate the mean of 'Error' and 'Error_sign'
df['Error'].mean()
df['Error_sign'].mean()

# Create a scatter plot to show 'Error' as a color-coded representation of the relationship between 'Expected' and 'Obtained'
plt.figure(figsize=(12, 6))
plt.scatter(vn1_data['Expected'], vn1_data['Obtained'], c=df['Error'], cmap='viridis')
plt.xlabel('Expected')
plt.ylabel('Obtained')
plt.title('Error vs Expected and Obtained for VN1')
plt.colorbar(label='Error')
plt.grid(True)
plt.show()
```
